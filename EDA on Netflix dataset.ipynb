## Movies vs TV Shows

#### Is there more Movies on Netflix than there are TV Shows or vice versa?

```python
netflix_movies_and_shows = df.groupby('type')['show_id'].count()
print(netflix_movies_and_shows)

netflix_movies_and_shows.plot(kind = 'bar', title = "Netflix Movies vs Shows", color = 'lightblue')
plt.xlabel('Type')
plt.xticks(rotation = 0)
plt.show()
```

- From the visualisation above, it is easy to deduce that there are twice as much movies on Netflix than there are TV Shows

### For easier analysis and understanding, I will have to segreggate the dataset into the two categories; Movies and TV Shows

```python
# Subsetting the dataset into two categories; Movies and Shows
netflix_movies = df[df['type'] == "Movie"]
netflix_shows = df[df['type'] == "TV Show"]
```

# Movies

#### First, let me use this dataset to answer questions like:

#### 1.Countries with the most movies on Netflix
#### 2.Directors that have directed the most movies on Netflix
#### 3.Number of movies on Netflix over the years
#### 4.Ratings of movies on Netflix

### 1.Countries with the most movies on Netflix

#### Unsurprisingly, we have the USA leading the pack with massive distance

```python
movie_top_countries = netflix_movies.groupby('country')['show_id'].count().sort_values(ascending = False).head(10)
print(movie_top_countries)
movie_top_countries.plot(kind = 'barh', title = 'Movies: Top 10 countries on Netflix', color = 'lightgreen')

plt.show()
```

### 2. Directors that have directed the most movies

#### Note: Considering that the "director" columnn is the one with over 2k null values, the result of this visualization might be deceptive.

#### You would think that directors from USA would be leading this chart but they are not. This might suggest more spread than directors from other countries

```python
top_directors = netflix_movies.groupby('director')['show_id'].count().sort_values(ascending = False).head(10)
print(top_directors)
top_directors.plot(kind = 'barh', title = 'Top Movie Directors on Netflix', color = 'lightpink')
plt.ylabel('Director', rotation = 0)
plt.show()
```

### 3. Number of movies on Netflix over the years

#### The number of movies on Netflix kept rising steadily from 2000, peaked in 2017 and 2018, after which there has been a sharp decline.

```python
nb_movies_year = netflix_movies.groupby('release_year')['show_id'].count()
print(nb_movies_year)
nb_movies_year.plot(kind = 'line', title = 'Number of Movies on Netflix over the Years', color = 'lightblue')
plt.xlabel('Year')
plt.show()
```

#### For a closer picture, lets see the number of movies on the platform in the last 10 years. It will be cool to know the year in the last decade have the most number of movies.

```python
nb_movies_10 = netflix_movies.query('release_year > 2010')
nb_movies = nb_movies_10.groupby('release_year')['show_id'].count()
nb_movies.plot(kind = 'bar', title = 'Movies: The year with the most movies in recent years?', color = 'lightpink')
plt.xlabel('Year')
plt.xticks(rotation = 45)
plt.show()
print(nb_movies.head(10))
```

### 4. Rating of movies on Netflix

#### There seems to be a movie for everyone on netflix

```python
rating = netflix_movies.groupby('rating')['show_id'].count().sort_values(ascending = False).head(10)
print(rating)
rating.plot(kind = 'bar', title = 'Rating of Movies on Netfix', color = 'lightblue')
plt.xticks(rotation = 45)
plt.xlabel('Rating')
plt.show()
```

# TV Shows

#### I will also answer the same set of questions I did with the movies;

#### 1.Countries with the most TV Shows on Netflix
#### 2.Directors that have directed the most TV shows on Netflix
#### 3.Number of TV Shows on Netflix over the years
#### 4.Ratings of TV Shows on Netflix

### 1. Countries with the most TV Shows on Netflix

#### Again ,USA leads the pack with UK coming 2nd this time around

```python
shows_top_countries = netflix_shows.groupby('country')['show_id'].count().sort_values(ascending = False).head(10)
print(shows_top_countries.head(10))
shows_top_countries.plot(kind = 'barh', title = 'Countries with most TV Shows on Netflix', color = 'orange')
plt.ylabel('Country', rotation = 0)
plt.show()
```

### 2. Directors that have directed the most TV shows on Netflix

#### Charlie Haskell, Koichi Sakamoto seems to the director that has directed the most TV shows on Netflx with 3 TV shows.

```python
show_directors = netflix_shows.groupby('director')['show_id'].count().sort_values(ascending = False).head(10)
print(show_directors)
show_directors.plot(kind = 'barh', title = 'Directors with most TV Shows on Netflix', color = 'yellow')
plt.ylabel('Directors', rotation = 90)
plt.show()
```

### 3. Number of TV Shows on Netflix over the years

#### From the visualisation below, there is little or no movement in this regard until 2000 and the number started increasing progressively and just like that of movies, a sharp dip in around 2019

```python
nb_shows = netflix_shows.groupby('release_year')['show_id'].count()
nb_shows.plot(kind = 'line', title = 'Number of TV Shows on Netflix over the years', color = 'orange')
plt.xlabel('Year')
plt.ylabel('Number of TV Shows')
plt.show()
```

```python
nb_shows =netflix_shows.query('release_year > 2010')
nb_shows_10 = nb_shows.groupby('release_year')['show_id'].count()
nb_shows_10.plot(kind = 'bar', title = 'TV Show: The year with the most TV Shows in recent years?', color = 'lightblue')
plt.xlabel('Year')
plt.xticks(rotation = 45)
plt.show()
print(nb_shows_10.head(10))
```

### 4. Ratings of TV Shows on Netflix

#### There is a show for everyone on Netflix although adults have more options than the rest of the age categories

```python
tv_shows_rating = netflix_shows.groupby('rating')['show_id'].count().sort_values(ascending = False)
print(tv_shows_rating)
tv_shows_rating.plot(kind = 'bar', title = 'Rating of TV Shows on Netflix', color = 'lightgreen')
plt.xticks(rotation = 0)
plt.ylabel('Number of TV Shows')
plt.xlabel('Ratings')
plt.show()
```
